## `survey_2017['CompensationAmount'][:10]` 的意思與功能

`survey_2017` 是一個 `DataFrame` 物件，從這個物件中我們可以分離出許多個變數欄位，例如 `survey_2017['CompensationAmount']` 就是將 `CompensationAmount`（年薪）從其中擷取出來，而 `survey_2017['CompensationAmount'][:10]` 是將前 10 個受訪者的年薪擷取出來的語法。

至於 `[:10]` 為什麼是從 0 到 9 而不是 1 到 10 呢？原因有二：

1. Python 的整數索引由 0 起始，而非我們慣常的由 1 起始；
2. Python 的慣例是包含起始、但不包含終止，因此 `[:10]` 表示由 0 起始、到 10 終止，包含 0 卻不包含 10。
